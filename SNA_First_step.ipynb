```python
import pandas as pd
import facebook_scraper
from facebook_scraper import *
set_user_agent("Mozilla/5.0 (compatible; Googlebot/2.1; +http://www.google.com/bot.html)")
# Parse the cookies from the cookies.txt file
cookies = '/content/sample_data/MYcookies.txt'

# Scrape the posts in the group
posts = get_posts(group='696067403864694' , cookies=cookies , options={"comments": True , "reactors": True , "progress": True})
# print(posts)
post_df_full = pd.DataFrame()

for post in posts:
    # print(post)
    post_entry = post
    fb_post_df = pd.DataFrame.from_dict(post_entry, orient='index')
    fb_post_df = fb_post_df.transpose()    
    post_df_full = post_df_full.append(fb_post_df)

```
